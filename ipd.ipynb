````python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

def downsample(filters, size, apply_batchnorm=True):
    initializer = tf.random_normal_initializer(0., 0.02)

    result = keras.Sequential()
    result.add(layers.Conv2D(filters, size, strides=2, padding='same',
                             kernel_initializer=initializer, use_bias=False))

    if apply_batchnorm:
        result.add(layers.BatchNormalization())

    result.add(layers.LeakyReLU())

    return result

def upsample(filters, size, apply_dropout=False):
    initializer = tf.random_normal_initializer(0., 0.02)

    result = keras.Sequential()
    result.add(layers.Conv2DTranspose(filters, size, strides=2,
                                       padding='same',
                                       kernel_initializer=initializer,
                                       use_bias=False))

    result.add(layers.BatchNormalization())

    if apply_dropout:
        result.add(layers.Dropout(0.5))

    result.add(layers.ReLU())

    return result

def UNet(input_shape=(128, 128, 3)):
    inputs = layers.Input(shape=input_shape)

    down_stack = [
        downsample(64, 4, apply_batchnorm=False),  # (bs, 64, 64, 64)
        downsample(128, 4),  # (bs, 32, 32, 128)
        downsample(256, 4),  # (bs, 16, 16, 256)
        downsample(512, 4),  # (bs, 8, 8, 512)
        downsample(512, 4),  # (bs, 4, 4, 512)
        downsample(512, 4),  # (bs, 2, 2, 512)
        downsample(512, 4),  # (bs, 1, 1, 512)
    ]

    up_stack = [
        upsample(512, 4, apply_dropout=True),  # (bs, 2, 2, 512)
        upsample(512, 4, apply_dropout=True),  # (bs, 4, 4, 512)
        upsample(512, 4, apply_dropout=True),  # (bs, 8, 8, 512)
        upsample(256, 4),  # (bs, 16, 16, 256)
        upsample(128, 4),  # (bs, 32, 32, 128)
        upsample(64, 4),  # (bs, 64, 64, 64)
    ]

    initializer = tf.random_normal_initializer(0., 0.02)
    last = layers.Conv2DTranspose(input_shape[-1], 4, strides=2,
                                  padding='same',
                                  kernel_initializer=initializer,
                                  activation='tanh')  # (bs, 128, 128, 3)

    x = inputs

    # Downsampling through the model
    skips = []
    for down in down_stack:
        x = down(x)
        skips.append(x)

    skips = skips[:-1] # Remove the last skip (bottleneck output)
    skips = reversed(skips)

    # Upsampling and establishing the skip connections
    for up, skip in zip(up_stack, skips):
        x = up(x)
        x = layers.Concatenate()([x, skip])

    x = last(x)

    return keras.Model(inputs=inputs, outputs=x)

model = UNet()
model.summary()
````

````python
T = 1000

beta = tf.cast(tf.linspace(0.0001, 0.02, T), tf.float32)
alpha = 1.0 - beta
alpha_bar = tf.math.cumprod(alpha, axis=0)

print("Diffusion parameters defined: ✅")
print(f"T: {T}")
print(f"Beta shape: {beta.shape}")
print(f"Alpha shape: {alpha.shape}")
print(f"Alpha_bar shape: {alpha_bar.shape}")
````

**Reasoning**:
The subtask requires defining a `forward_diffusion` function that adds Gaussian noise to a clean image at a specific timestep. I will implement this function according to the instructions, leveraging TensorFlow for all operations and ensuring it is decorated with `@tf.function` for performance.



````python
import tensorflow as tf

@tf.function
def forward_diffusion(x0, t):
    # Retrieve alpha_bar_t for the given timestep t (0-indexed)
    alpha_bar_t = tf.gather(alpha_bar, t - 1)

    # Reshape alpha_bar_t to match image dimensions for broadcasting
    alpha_bar_t_reshaped = tf.reshape(alpha_bar_t, [tf.shape(x0)[0], 1, 1, 1])

    # Generate random Gaussian noise with the same shape as x0
    noise = tf.random.normal(tf.shape(x0))

    # Calculate the noisy image xt at timestep t
    xt = tf.sqrt(alpha_bar_t_reshaped) * x0 + tf.sqrt(1 - alpha_bar_t_reshaped) * noise

    return xt, noise

print("Forward diffusion function defined ✅")
````

**Reasoning**:
I need to define the `train_step` function, which is the core of the diffusion model's training process. This function will handle sampling clean images, adding noise, predicting noise with the U-Net, calculating the loss, and applying gradients to update the model, as described in the subtask.



````python
import tensorflow as tf

@tf.function
def train_step(batch_x0):
    batch_size = tf.shape(batch_x0)[0]

    # 2. Randomly sample timesteps (t) for each image
    # tf.random.uniform generates floats, so cast to int32 and ensure range [1, T]
    t = tf.random.uniform(shape=[batch_size], minval=1, maxval=T + 1, dtype=tf.int32)

    with tf.GradientTape() as tape:
        # 3. Add noise to the clean images using the `forward_diffusion` function
        xt, epsilon = forward_diffusion(batch_x0, t)

        # 4. Feed the noisy images and timesteps to the `UNet` model to predict the noise
        # U-Net takes noisy image and timestep as input
        predicted_noise = model([xt, t])

        # 5. Compute the loss between the predicted noise and the actual noise
        loss = loss_fn(epsilon, predicted_noise)

    # 6. Compute the gradients of the `loss` with respect to the `model.trainable_variables`
    gradients = tape.gradient(loss, model.trainable_variables)

    # 7. Apply the computed gradients to update the model's weights
    optimizer.apply_gradients(zip(gradients, model.trainable_variables))

    return loss

print("Train step function defined ✅")
````

````python
import tensorflow as tf
import os
import cv2
import numpy as np

# Define paths and IMG_SIZE
base = "/content/drive/MyDrive/lunar_train_dataset/"
low_path = base + "low/"
high_path = base + "high/"
IMG_SIZE = 128

# Define load_images function
def load_images(path):
    images = []
    files = sorted(os.listdir(path))

    for file in files:
        img = cv2.imread(os.path.join(path, file))
        if img is None:
            continue
        img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        img = img / 255.0
        images.append(img)

    return np.array(images)

# Load data
train_low = load_images(low_path)
train_high = load_images(high_path)

BATCH_SIZE = 8

dataset = tf.data.Dataset.from_tensor_slices((train_low, train_high))
dataset = dataset.shuffle(1000).batch(BATCH_SIZE)

EPOCHS = 50

training_losses = []

print("Starting training loop...\n")

for epoch in range(EPOCHS):
    total_loss = 0.0
    num_batches = 0
    for batch_low, batch_high in dataset: # dataset yields (train_low, train_high)
        # Pass batch_high (clean image) to train_step as batch_x0
        loss = train_step(batch_high)
        total_loss += loss
        num_batches += 1

    avg_loss = total_loss / num_batches
    training_losses.append(avg_loss)

    print(f"Epoch {epoch+1}/{EPOCHS}, Loss: {avg_loss:.4f}")

print("\nTraining complete! ✅")
````

**Reasoning**:
The subtask requires implementing the `reverse_diffusion` function. This involves defining the function, initializing variables, pre-calculating diffusion parameters like `sqrt_recip_alpha_t`, `sqrt_recip_alpha_m_t`, and `posterior_variance_t`, and then iterating backwards through timesteps to iteratively denoise the image using the U-Net model. The calculations for `posterior_variance_t` need to carefully handle `alpha_bar_prev`.



````python
import tensorflow as tf

@tf.function
def reverse_diffusion(noisy_image):
    x = noisy_image

    # TensorFlow constants
    one = tf.constant(1.0, dtype=tf.float32)
    minus_one = tf.constant(-1.0, dtype=tf.float32)
    half = tf.constant(0.5, dtype=tf.float32)

    # Pre-calculate common terms
    sqrt_recip_alpha = tf.sqrt(one / alpha)
    sqrt_recip_alpha_m_t = tf.sqrt(one / alpha - one)

    # Calculate alpha_bar_prev for posterior variance calculation
    # alpha_bar_0 is 1.0
    alpha_bar_prev = tf.concat([[one], alpha_bar[:-1]], axis=0)

    # Calculate posterior variance (beta_t * (1 - alpha_bar_{t-1})) / (1 - alpha_bar_t)
    # The last value of alpha_bar is 0 when t = 0 so we need to handle it. Set to 0.
    posterior_variance = tf.where(
        alpha_bar == 0,
        tf.zeros_like(beta),
        beta * (one - alpha_bar_prev) / (one - alpha_bar)
    )
    # For t=1, posterior_variance can be set to beta[0]
    posterior_variance = tf.tensor_scatter_nd_update(
        posterior_variance, [[0]], [beta[0]]
    )

    # Loop backwards from T-1 down to 0
    for t_idx in tf.range(T - 1, minus_one, minus_one):
        t = t_idx + 1 # Timestep t goes from T down to 1

        # Reshape for broadcasting
        t_tensor = tf.cast(t, dtype=tf.int32)
        # For each t, we need the corresponding scalar value from the precomputed tensors
        # alpha_bar, beta, sqrt_recip_alpha, sqrt_recip_alpha_m_t, posterior_variance
        # tf.gather requires batch dimension for t_tensor if t_tensor is a single scalar.
        # Instead, we can use indexing directly on the precomputed tensors.

        # Get scalar values for current timestep
        alpha_t = alpha[t_idx]
        sqrt_recip_alpha_t_scalar = sqrt_recip_alpha[t_idx]
        sqrt_recip_alpha_m_t_scalar = sqrt_recip_alpha_m_t[t_idx]
        posterior_variance_t_scalar = posterior_variance[t_idx]

        # Reshape scalars to match image dimensions for broadcasting [batch, 1, 1, 1]
        # The batch dimension is handled implicitly by tf.shape(x)[0] inside tf.reshape if x is batch_sized.
        # If x is single image, batch_size is 1.
        batch_size = tf.shape(x)[0]

        reshaped_sqrt_recip_alpha_t = tf.reshape(sqrt_recip_alpha_t_scalar, [batch_size, 1, 1, 1])
        reshaped_sqrt_recip_alpha_m_t = tf.reshape(sqrt_recip_alpha_m_t_scalar, [batch_size, 1, 1, 1])
        reshaped_posterior_variance_t = tf.reshape(posterior_variance_t_scalar, [batch_size, 1, 1, 1])

        # Predict noise using the U-Net model
        # The model expects a batch, so we need to ensure t_tensor is broadcastable or replicated.
        t_batch = tf.cast(tf.fill([batch_size], t), dtype=tf.int32)
        predicted_noise = model([x, t_batch])

        # Compute the mean for denoising
        mean = (reshaped_sqrt_recip_alpha_t * (x - reshaped_sqrt_recip_alpha_m_t * predicted_noise))

        # If t > 1, add new noise; otherwise, it's the final denoised image without additional noise
        if t > 1:
            noise = tf.random.normal(tf.shape(x))
            x = mean + tf.sqrt(reshaped_posterior_variance_t) * noise
        else: # t == 1, no extra noise added in the last step
            x = mean

    # Clip the final image values to [0.0, 1.0]
    x = tf.clip_by_value(x, 0.0, 1.0)

    return x

print("Reverse diffusion function defined ✅")
````

## Define PSNR Function

### Subtask:
Implement a Python function `calculate_psnr` that takes two TensorFlow tensors (ground truth and predicted image) as input and returns the Peak Signal-to-Noise Ratio (PSNR) using TensorFlow operations. Ensure it handles image value ranges correctly (0-1).


**Reasoning**:

````python
import tensorflow as tf

def calculate_psnr(ground_truth, predicted_image):
    """
    Calculates the Peak Signal-to-Noise Ratio (PSNR) between two images.

    Args:
        ground_truth (tf.Tensor): The ground truth image tensor (values 0-1).
        predicted_image (tf.Tensor): The predicted image tensor (values 0-1).

    Returns:
        tf.Tensor: The PSNR value.
    """
    # Ensure both tensors are float32 for tf.image.psnr
    ground_truth = tf.cast(ground_truth, tf.float32)
    predicted_image = tf.cast(predicted_image, tf.float32)

    psnr_value = tf.image.psnr(ground_truth, predicted_image, max_val=1.0)
    return psnr_value

print("PSNR calculation function defined ✅")
````

````python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np

# Re-define diffusion parameters (from cell 71b50ab6) to ensure they are in scope
T = 1000
beta = tf.cast(tf.linspace(0.0001, 0.02, T), tf.float32)
alpha = 1.0 - beta
alpha_bar = tf.math.cumprod(alpha, axis=0)

# Helper function for timestep embedding, now as a Keras Layer
class TimestepEmbedding(layers.Layer):
    def __init__(self, dim, max_period=10000, **kwargs):
        super().__init__(**kwargs)
        self.dim = dim
        self.max_period = max_period
        self.half_dim = dim // 2

    def call(self, timesteps):
        freqs = tf.exp(
            tf.math.log(tf.cast(self.max_period, tf.float32)) *
            tf.range(start=0, limit=self.half_dim, dtype=tf.float32) /
            tf.cast(self.half_dim - 1, tf.float32)
        )
        args = tf.cast(timesteps, dtype=tf.float32) * freqs
        embedding = tf.concat([tf.sin(args), tf.cos(args)], axis=-1)
        return embedding

# Re-define UNet architecture and instantiate model (from cell 76bfb001) to ensure it is in scope
def downsample(filters, size, apply_batchnorm=True):
    initializer = tf.random_normal_initializer(0., 0.02)

    result = keras.Sequential()
    result.add(layers.Conv2D(filters, size, strides=2, padding='same',
                             kernel_initializer=initializer, use_bias=False))

    if apply_batchnorm:
        result.add(layers.BatchNormalization())

    result.add(layers.LeakyReLU())

    return result

def upsample(filters, size, apply_dropout=False):
    initializer = tf.random_normal_initializer(0., 0.02)

    result = keras.Sequential()
    result.add(layers.Conv2DTranspose(filters, size, strides=2,
                                       padding='same',
                                       kernel_initializer=initializer,
                                       use_bias=False))

    result.add(layers.BatchNormalization())

    if apply_dropout:
        result.add(layers.Dropout(0.5))

    result.add(layers.ReLU())

    return result

def UNet(input_shape=(128, 128, 3)):
    inputs_image = layers.Input(shape=input_shape, name='input_image')
    inputs_timestep = layers.Input(shape=(1,), dtype=tf.int32, name='input_timestep') # Timestep as a scalar for each batch item

    # Embed timestep using the new Layer
    t_emb = TimestepEmbedding(dim=512)(inputs_timestep) # Embed into 512-dim vector for bottleneck
    t_emb = layers.Dense(512, activation='relu')(t_emb)
    t_emb = layers.Dense(512)(t_emb) # Final embedding for the bottleneck
    t_emb = layers.Reshape((1, 1, 512))(t_emb) # Reshape to be broadcastable to bottleneck feature map

    x = inputs_image

    down_stack = [
        downsample(64, 4, apply_batchnorm=False),  # (bs, 64, 64, 64)
        downsample(128, 4),  # (bs, 32, 32, 128)
        downsample(256, 4),  # (bs, 16, 16, 256)
        downsample(512, 4),  # (bs, 8, 8, 512)
        downsample(512, 4),  # (bs, 4, 4, 512)
        downsample(512, 4),  # (bs, 2, 2, 512)
        downsample(512, 4),  # (bs, 1, 1, 512)
    ]

    up_stack = [
        upsample(512, 4, apply_dropout=True),  # (bs, 2, 2, 512)
        upsample(512, 4, apply_dropout=True),  # (bs, 4, 4, 512)
        upsample(512, 4, apply_dropout=True),  # (bs, 8, 8, 512)
        upsample(256, 4),  # (bs, 16, 16, 256)
        upsample(128, 4),  # (bs, 32, 32, 128)
        upsample(64, 4),  # (bs, 64, 64, 64)
    ]

    # Downsampling through the model
    skips = []
    for down in down_stack:
        x = down(x)
        skips.append(x)

    # Bottleneck: Add timestep embedding to the bottleneck feature map
    x = x + t_emb # Add timestep conditioning

    skips = skips[:-1] # Remove the last skip (bottleneck output)
    skips = reversed(skips)

    # Upsampling and establishing the skip connections
    for up, skip in zip(up_stack, skips):
        x = up(x)
        x = layers.Concatenate()([x, skip])

    initializer = tf.random_normal_initializer(0., 0.02)
    last = layers.Conv2DTranspose(input_shape[-1], 4, strides=2,
                                  padding='same',
                                  kernel_initializer=initializer,
                                  activation='tanh')  # (bs, 128, 128, 3)
    x = last(x)

    # Return model with two inputs
    return keras.Model(inputs=[inputs_image, inputs_timestep], outputs=x)

model = UNet() # Instantiate the model

# Re-define reverse_diffusion function (from cell db328520) to ensure it is in scope
@tf.function
def reverse_diffusion(noisy_image):
    x = tf.cast(noisy_image, tf.float32) # Explicitly cast to float32

    # TensorFlow constants (ensure integer constants for tf.range)
    one_float = tf.constant(1.0, dtype=tf.float32)
    minus_one_float = tf.constant(-1.0, dtype=tf.float32)
    one_int = tf.constant(1, dtype=tf.int32)
    minus_one_int = tf.constant(-1, dtype=tf.int32)
    zero_int = tf.constant(0, dtype=tf.int32)

    # Pre-calculate common terms
    sqrt_recip_alpha = tf.sqrt(one_float / alpha)
    sqrt_recip_alpha_m_t = tf.sqrt(one_float / alpha - one_float)

    # Calculate alpha_bar_prev for posterior variance calculation
    alpha_bar_prev = tf.concat([[one_float], alpha_bar[:-1]], axis=0)

    # Calculate posterior variance (beta_t * (1 - alpha_bar_{t-1})) / (1 - alpha_bar_t)
    # The last value of alpha_bar is 0 when t = 0 so we need to handle it. Set to 0.
    posterior_variance = tf.where(
        alpha_bar == 0,
        tf.zeros_like(beta),
        beta * (one_float - alpha_bar_prev) / (one_float - alpha_bar)
    )
    # For t=1, posterior_variance can be set to beta[0]
    posterior_variance = tf.tensor_scatter_nd_update(
        posterior_variance, [[zero_int]], [beta[zero_int]]
    )

    # Loop backwards from T-1 down to 0, ensuring t_idx is tf.int32
    for t_idx_float in tf.range(T - one_int, minus_one_int, minus_one_int, dtype=tf.int32):
        t_idx = tf.cast(t_idx_float, tf.int32) # Cast to int32 for indexing
        t = t_idx + one_int # Timestep t goes from T down to 1

        # Get scalar values for current timestep
        alpha_t = alpha[t_idx]
        sqrt_recip_alpha_t_scalar = sqrt_recip_alpha[t_idx]
        sqrt_recip_alpha_m_t_scalar = sqrt_recip_alpha_m_t[t_idx]
        posterior_variance_t_scalar = posterior_variance[t_idx]

        # Reshape scalars to match image dimensions for broadcasting [batch, 1, 1, 1]
        batch_size = tf.shape(x)[0]

        reshaped_sqrt_recip_alpha_t = tf.reshape(sqrt_recip_alpha_t_scalar, [batch_size, 1, 1, 1])
        reshaped_sqrt_recip_alpha_m_t = tf.reshape(sqrt_recip_alpha_m_t_scalar, [batch_size, 1, 1, 1])
        reshaped_posterior_variance_t = tf.reshape(posterior_variance_t_scalar, [batch_size, 1, 1, 1])

        # Predict noise using the U-Net model
        # t_batch needs to be a tensor with shape (batch_size, 1) or (batch_size,) if model expects scalar per batch item
        t_batch_model_input = tf.cast(tf.fill([batch_size, 1], t), dtype=tf.int32)
        predicted_noise = model([x, t_batch_model_input])

        # Compute the mean for denoising
        mean = (reshaped_sqrt_recip_alpha_t * (x - reshaped_sqrt_recip_alpha_m_t * predicted_noise))

        # If t > 1, add new noise; otherwise, it's the final denoised image without additional noise
        if t > one_int:
            noise = tf.random.normal(tf.shape(x))
            x = mean + tf.sqrt(reshaped_posterior_variance_t) * noise
        else: # t == 1, no extra noise added in the last step
            x = mean

    # Clip the final image values to [0.0, 1.0]
    x = tf.clip_by_value(x, 0.0, 1.0)

    return x

denoised_images = []

print("Generating denoised images...")

for i, low_light_img in enumerate(test_low):
    # Expand dimensions to create a batch of size 1 for the reverse_diffusion function
    batched_low_light_img = tf.expand_dims(low_light_img, axis=0)

    # Generate the denoised image using reverse_diffusion
    denoised_img_tensor = reverse_diffusion(batched_low_light_img)

    # Remove the batch dimension and convert to numpy array
    denoised_img_np = denoised_img_tensor[0].numpy()
    denoised_images.append(denoised_img_np)

# Convert the list of denoised images to a NumPy array
denoised_images = np.array(denoised_images)

print(f"Generated {len(denoised_images)} denoised images. Shape: {denoised_images.shape} ✅")
````
